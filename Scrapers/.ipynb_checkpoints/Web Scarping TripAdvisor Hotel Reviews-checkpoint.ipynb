```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys#may not used
import urllib.parse
import urllib.error
from bs4 import BeautifulSoup
import requests
import re
import time
import pandas as pd
```

```python
df_hotels = pd.read_csv('hotel_features.csv')
```

```python
def convert_str_int(s):
    remove = s.replace(",","")
    return int(remove.split(' ')[0])
```

```python
df_hotels['number_reviews'] = df_hotels['total_reviews'].apply(convert_str_int)
```

```python
df_most_reviews = df_hotels[df_hotels['number_reviews'] > 1000]
```

```python
# sorted by total_reviews
df_sorted = df_most_reviews.sort_values(by = ['total_reviews'] ,ascending = False)
```

```python
# removing duplicates in sorted df
df_sorted['hotel_names'].nunique()
```

```python
df_sorted.drop([368 , 37] , inplace = True)
```

```python
# removing duplicates in sorted df
df_sorted.drop_duplicates(subset = 'hotel_names', inplace = True)
# df_sorted.head(10)
```

```python
list_urls = df_sorted['url'].tolist()[:5]
list_urls
```

```python
list_urls = df_sorted['url'].tolist()[6:11]
list_urls
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.90 Safari/537.36'}
list_reviews = []
list_rating = []
list_users = []
for url in list_urls: 
    find_numbers = re.findall(r'[0-9]+', url)
    find_name_hotel = re.findall('Reviews-.*', url)

    for num_pages in range(0,2505,5): # dos paginas cada hotel
            list_url = url[:50] + find_numbers[1] + '-Reviews-' + 'or' + str(num_pages) + find_name_hotel[0][7:]
            driver.get(list_url)
            time.sleep(5)
            element_list = driver.find_elements_by_xpath("//span[@class='taLnk ulBlueLinks']")
            for ele in element_list:
                try:
                        ele.click()
                except:
                    pass
            html = driver.page_source
            response = requests.get(list_url, headers=headers, verify=False).text
            soup_1 = BeautifulSoup(response, "lxml")
            x = soup_1.find_all('q', class_="location-review-review-list-parts-ExpandableReview__reviewText--gOmRC")
            for r in x:
                list_reviews.append(r.text)
                
            rating_find_in_soup = soup_1.find_all('div', 'location-review-review-list-parts-RatingLine__bubbles--GcJvM')
            rating_string = str(rating_find_in_soup) # converted to srt to do regex
            rating =  re.findall(r'[0-9]{1,2}',rating_string)
            for rate in rating:
                list_rating.append(rate)

            
            users = driver.find_elements_by_xpath("//a[@class='ui_header_link social-member-event-MemberEventOnObjectBlock__member--35-jC']")
            for user in users:
                list_users.append(user.text)

df = pd.DataFrame([list_reviews, list_rating, list_users]).transpose()
df.columns = ['reviews', 'review_rating', 'user']
```
